2. Изначально БД находилась во 2-й нормальной форме, так как у каждой таблицы был первичный ключ. Я привет БД к 3 НФ, убрав из таблиц транзитивность. Для этого из таблицы transaction я выделил в отдельную таблицу данные о продукте (таблица product на схеме), а из таблицы customer я выделил данные о работе клиента и его адрсе (таблицы job и address на схеме соответственно).

3. Далее я создал таблицы по своей схеме с помощью DBeaver:

```python
create table transaction (
    transaction_id int primary key,
    product_id int,
    customer_id int,
    transaction_date timestamp,
    online_order bool,
    order_status varchar(20)
);

create table customer (
  customer_id int primary key,
  first_name text,
  last_name text,
  gender varchar(20),
  DOB timestamp,
  job_id int,
  address_id int,
  wealth_segment text,
  deceased_indicator varchar(10),
  owns_car varchar(10),
  property_valuation int
);

create table job (
  job_id int primary key,
  job_title text,
  job_industry_category text
);

create table product (
  product_id int primary key,
  brand text,
  product_line text,
  product_class text,
  product_size text,
  list_price float4,
  standard_cost float4
);

create table address (
  address_id int primary key,
  address text,
  postcode int,
  state text,
  country text
);
```
